### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['blue', 'red', 'green', 'orange']

# Ensure data is valid
if len(labels) == len(sizes) == len(colors):
    try:
        # Create pie chart
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})

        # Customization
        plt.title('Programming Language Usage')

        # Show the plot
        plt.show()

    except Exception as e:
        print(f"An error occurred: {e}")

else:
    print("Error: The lengths of labels, sizes, and colors must be the same.")
```
